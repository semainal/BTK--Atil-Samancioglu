```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,60,70,80,90]
kiloListesi = [30,40,50,60,70,80,70,60,75]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "g")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Grafiği")
```

```python
numpyDizisi1 = np.linspace (1,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "r")
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "r--")
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"g*-")
plt.subplot(1,2,2) # 1 sıra, 2 grafik, 2.grafiği çiziyorum.
plt.plot(numpyDizisi2,numpyDizisi1, "r--")
```

# figure

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.5,0.5])
figureAxes.plot(numpyDizisi1,numpyDizisi2, "b")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, "r")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1, "g")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Alt Grafik Başlığı")
```

# subplots

```python
(benimFigure,benimEksen) = plt.subplots(nrows=1, ncols=2)
type(benimEksen)

for eksen in benimEksen:
    eksen.plot(numpyDizisi1,numpyDizisi2, "r")
    eksen.set_xlabel("X Ekseni")
    
plt.tight_layout()
```

# görsel geliştirme

```python
yeniFigure = plt.figure()
yeniEksen = yeniFigure.add_axes([0.3,0.3,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi2, label = "Numpy Dizisi **2")
yeniEksen.legend()
```

# 2 grafik çizmek için

```python
newFigure = plt.figure()
newEksen = newFigure.add_axes([0.1,0.1,0.9,0.9])

newEksen.plot(numpyDizisi1, numpyDizisi1 **2, label= "Numpy Dizisi**2")
newEksen.plot(numpyDizisi1,numpyDizisi1**3, label = "Numpy Dizis**3")
newEksen.legend()
```

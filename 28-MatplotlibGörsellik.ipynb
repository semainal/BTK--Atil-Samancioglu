```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(1,10,20)
numpyDizisi2 = numpyDizisi1 **2
```

```python
(benimFifure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2, color= "#50145A", alpha= 0.3)
benimEksen.plot(numpyDizisi2,numpyDizisi1, color= "#88DBBA")
```

# kalınlık & stiller

```python
(benimFigure, benimEksen) =  plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color = "blue", linewidth = 3.0)
benimEksen.plot(numpyDizisi1, numpyDizisi1+ 3, color= "red", linewidth = 3.0)
benimEksen.plot(numpyDizisi1, numpyDizisi1+ 4, color= "green", linewidth = 3.0, linestyle= "--")
benimEksen.plot(numpyDizisi1, numpyDizisi1+ 5, color= "green", linewidth = 3.0, linestyle = ":")
benimEksen.plot(numpyDizisi1, numpyDizisi1+ 6, color= "green", linewidth = 3.0, linestyle = "-.")
benimEksen.plot(numpyDizisi1, numpyDizisi1+ 7, color= "#000000", linewidth = 3.0, linestyle = "--", marker = "o", markersize= 8, markerfacecolor = "red")
```
